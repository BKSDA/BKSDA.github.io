```python
# Drop columns that we don't want for the analysis
trips_final_df = trips_final.drop(['start_station_id', 'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng'], axis=1)
```

```python
#Changing the data type from object to datetime for the started_at column
trips_final_df.started_at = pd.to_datetime(trips_final_df.started_at, format='mixed')
```

```python
#Changing the data type from object to datetime for the ended_at column
trips_final_df.ended_at = pd.to_datetime(trips_final_df.ended_at, format='mixed')
```

```python
trips_final_df.info()
```

```python
# Creating a column for ride_length = ended_at - started_at
trips_final_df['ride_length'] = trips_final_df['ended_at'] - trips_final_df['started_at']
trips_final_df
```

```python
# Create a column for weekday using the started_at column as the date
trips_final_df['weekday'] = trips_final_df['started_at'].dt.day_name()
trips_final_df
```

```python
#Dropping all Null values from the analysis
trips_final_df = trips_final_df.dropna()
```

```python
#Adding a column 'month' using the date of started_at column. This way we can analyze with graphs later monthly rides.
trips_final_df['month'] = pd.DataFrame({
    'started_at': pd.date_range('2022-05-01', '2023-04-01', freq='1M')
})
trips_final_df['month'] = trips_final_df.started_at.dt.month_name()
```

```python
# Finally I reset the index with drop = True, this way the rows count from 0,1,2,3,.... in order instead of having the wrong row number name.
trips_final_df = trips_final_df.reset_index(drop=True).copy()
trips_final_df
```

```python
#In which months/days were the ride lengths below average?
#If the value is less than the average ride length per month, highlight the value in blue.

avg_ride = trips_final_df.groupby(['weekday', 'month']).count()['ride_length'].mean()
print("The average ride length is " + str(avg_ride))

member_casual_day1 = trips_final_df.groupby(['month', 'weekday']).count()['ride_length'].unstack()

def style_negative(v, props=''):
    return props if v < avg_ride else None
s2 = member_casual_day1.style.applymap(style_negative, props='color:blue;')\
              .applymap(lambda v: 'opacity: 20%;' if (v < 0.3) and (v > -0.3) else None)
s2

#Highlight the maximum in a column in darkgreen

def highlight_max(s, props=''):
    return np.where(s==np.nanmax(s.values), props,'')
s2.apply(highlight_max,props='color:white;background-color:darkgreen',axis=0)
```
